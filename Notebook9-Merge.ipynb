# Unir dataframes usando *merge()*

````python
import pandas as pd
````

Creamos los dataframes con los que trabajaremos

````python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'Mix':['y2','y3','a2','a3']},
    index = ['y2','y3','a2','a3'])

df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
````

````python
df_left
````

````python
df_right
````

Ahora los unimos

````python
pd.merge(df_left,df_right)
````

Notemos que se unen usando la columna que tienen en común. También se pude modificar el tipo de unión (merge), en este caso es de tipo:

````python
pd.merge(df_left,df_right, how = 'inner', on = 'Y')
````


También se puede especificar mediante qué columna se van a unir

````python
pd.merge(df_left,df_right, how = 'inner', left_on = 'Mix', right_on = 'Y')
````

````python
pd.merge(df_left,df_right, how = 'inner', left_on = 'Mix', right_on = 'A')
````

Por ejemplo, una unión de derecha es:

````python
pd.merge(df_left,df_right, how = 'right', on = 'Y')
````

Y una unión "outer"

````python
pd.merge(df_left,df_right, how = 'outer', on = 'Y')
````

Para el siguiente ejemplo, modifiquemos la df_left con una columna A en lugar de Mix

````python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'A':['a0','a1','a2','a3']})
df_left
````

````python
pd.merge(df_left,df_right, how = 'outer', on = ['Y', 'A'])
````

````python
pd.merge(df_left,df_right, how = 'outer', on = 'A')
````

En esta última unión notemos que las Y tienen sufijos, éstos se pueden modificar

````python
pd.merge(df_left,df_right, how = 'outer', on = 'A', suffixes = ['_left', '_right'])
````
