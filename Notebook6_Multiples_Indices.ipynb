Ahora pivoteamos las filas para convertirlas en columnas, y ordenamos los datos primero por "year" y luego por "Country", de etsa manera obtenemos el mismo dataset de la clase

# Aquí termina la preparación del dataframe
Notemos que ahora trabajaremos con df en lugar de df_pob

Editamos el formato de decimales

```python
pd.options.display.float_format = '{:,.1f}'.format
df
```

Para el uso de múltiples índices, convertimos "year" a variable categórica y tipo texto (string)

```python
df['year'] = pd.Categorical(df['year'].apply(str))
df
```

```python
df['year'] = pd.Categorical(df['year'].apply(str))
```

Comprobamos las variables de la tabla

```python
df.dtypes
```

Reducimos la información, de modo que solo se muestren algunos países. Para ello, primero creamos un filtro

```python
filtro = df['Country Name'].isin(['Mexico', 'Colombia'])
filtro
```

Luego creamos un dataframe al que se aplicará el filtro

```python
df_filtro = df[filtro]
df_filtro
```

Ahora creamos los índices múltiples, de modo que "Country Name" y "year" sean los nuevos índices

```python
df_filtro.set_index(['Country Name', 'year'])
```

Ahora lo ordenamos con la función *sort_index()* y lo guardamos en el dataframe

```python
df_filtro = df_filtro.set_index(['Country Name', 'year']).sort_index()
df_filtro
```

Los datos han quedado separados y esto permite un mejor análisis

Se pude mostrar solo un país

```python
df_filtro.loc['Mexico']
```

Y se puede mostrar solo un año

```python
df_filtro.loc['Mexico'].loc['2016',:]
```

Otra forma para hacer lo anterior es que en lugar de usar *loc[]* podemos usar *xs()*

```python
df_filtro.xs(['Mexico'])
```

```python
df_filtro.xs(['Mexico','2018'])
```

```python
df_filtro.xs('2018', level = 'year')
```

Podemos organizar toda la tabla con índices

```python
df_countries = df.set_index(['Country Name', 'year']).sort_index(ascending = [True, True])
df_countries
```

Se puede filtrar de un país a otro en orden alfabético y entre años

```python
ids = pd.IndexSlice
df_countries.loc[ids['Aruba':'Austria','2015':'2017'],:].sort_index()
```

Se puede ver la composición del dataframe

```python
df_countries.index
```

Lo anterior se puede hacer en un primer nivel

```python
df_countries.index.get_level_values(0)
```

Por otro lado, podemos visualizar la tabla de otra manera con

```python
df_countries['pob']
```

Y a esto se le pueden aplicar reglas de filtrado como anteriormente, pero de manera más fácil

```python
df_countries['pob']['Mexico']['2015']
```

Todo lo anterior tiene importancia a la hora de aplicar funciones sobre la tabla, por ejemplo, para sumar la poblanción de todos los países para cada año (población mundial por año), hacemos:

```python
df_countries.sum(level = 'year')
```

Los múltiples índices también permiten reordenar los datos, por ejemplo:

```python
df_filtro.unstack('year')
```
